**№3**  
Пришла информация, что на сервере в Москве возник сбой и баланс за отправленные смс не списался (у пользователей из Москвы). Спишите с каждого пользователя из Москвы деньги за смс. Количество отправленных смс за которые не списали деньги в столбце sms. Стоимость 1 смс - 3 рубля. (обновите колонку balance)

```python
df.head()
```

```python
df['balance'] = df['balance'].sub(3 * df[df.city == 'Moscow'].sms, fill_value=0).astype(int)
df.head()
```

**№5**  
На совещании было принято решение детальнее проработать вопрос списания денежных средств. Теперь, спишите деньги (у пользователей из Москвы за смс) с учетом тарифной сетки:
ВИП - 1 рубль за СМС
Все за 300 - 2 рубля за СМС
Всё за 500 - 3 рубля за СМС
Всё за 800 - 4 рубля за СМС
Промо - 7 рублей за СМС
Корпоративный - 0 рублей за СМС
Всё за 250 (архив) - 5 рублей за СМС

```python
df.head()
```

```python
dct_sms_cost = pd.Series({
        'ВИП': 1,
        'Все за 300': 2,
        'Всё за 500': 3,
        'Всё за 800': 4,
        'Промо': 7,
        'Корпоративный': 0,
        'Всё за 250 (архив)': 5}, name='price').to_frame()
dct_sms_cost
```

```python
pd.merge(df, dct_sms_cost, left_on='tarif', right_index=True)
```

```python
df.loc[df.city == 'Moscow', 'balance'] = df.balance - df.sms * pd.merge(df, dct_sms_cost, left_on='tarif', right_index=True).price
```
