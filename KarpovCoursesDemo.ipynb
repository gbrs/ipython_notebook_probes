```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
taxi = pd.read_csv('https://stepik.org/media/attachments/lesson/360340/2_taxi_nyc.csv')
```

```python
taxi.columns = [x.replace(' ', '_') for x in taxi.columns]
```

```python
taxi
```

```python
pickups_by_mon_bor = taxi.groupby(['borough', 'pickup_month'])\
                            .agg({'pickups': sum})\
                            .sort_values(by='pickups', ascending=False)\
                            .reset_index()
```

```python
def temp_to_celcius(s):
    return (s - 32) * 5 / 9
#     return s.apply(lambda x: (x - 32) * 5 / 9)
```

```python
taxi['temp'][:5]
```

```python
taxi['temp_C'] = temp_to_celcius(taxi['temp'])
taxi['temp_C'][:5]
```

```python
bookings = pd.read_csv('https://stepik.org/media/attachments/lesson/360344/bookings.csv', sep=';')
bookings.columns = bookings.columns.str.replace(' ', '_').str.lower()
```

```python
bookings.head()
```

```python
bookings_head = bookings.head(7)
```

```python
bookings.dtypes.value_counts()
```

```python
bookings.query('is_canceled == 0') \
        .groupby('country') \
        .agg({'is_canceled': 'count'}) \
        .sort_values('is_canceled', ascending=False) \
        .head()
```

```python
bookings.groupby('hotel') \
        .agg({'stays_total_nights': 'mean'})
```

```python
bookings.query('assigned_room_type != reserved_room_type').shape[0]
```

```python
bookings.query('(is_canceled == 1) & (hotel == "City Hotel")') \
        .groupby('arrival_date_year') \
        .arrival_date_month \
        .value_counts() \
        .sort_values(ascending=False)
```

```python
bookings.loc[:, ['adults', 'children', 'babies']].mean()
```

```python
bookings['total_kids'] = bookings.children + bookings.babies
```

```python
bookings.groupby('hotel').total_kids.mean().round(2)
```
