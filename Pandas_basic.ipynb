ВИП - 1 рубль за СМС
Все за 300 - 2 рубля за СМС

Промо - 7 рублей за СМС

```python
dct_sms_cost = pd.Series({
        'ВИП': 1,
        'Все за 300': 2,
        'Всё за 500': 3,
        'Всё за 800': 4,
        'Промо': 7,
        'Корпоративный': 0,
        'Всё за 250 (архив)': 5}, name='price').to_frame()
dct_sms_cost
```

```python
pd.merge(df, dct_sms_cost, left_on='tarif', right_index=True)
```

```python
df.loc[df.city == 'Moscow', 'balance'] \
            = df.balance \
            - df.sms * pd.merge(
                    df
                    , dct_sms_cost
                    , left_on='tarif'
                    , right_index=True)\
                .price
df
```
